# 01 - Valuation por Multiplos

Coleta fundamentos de empresas via `yfinance` e calcula multiplos de mercado:
**P/E**, **P/B**, **EV/EBITDA** — comparando contra peers para gerar uma decisao de valuation relativo.

**Decisao gerada:** *esta cara ou barata vs. peers?*

```python
import sys
sys.path.append('../src')

from data_fetcher import get_fundamentals
from valuation import calcular_multiplos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# Definir peer group
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'META', 'AMZN']

print('Coletando fundamentos...')
df_fund = get_fundamentals(TICKERS)
df_fund[['shortName', 'sector', 'marketCap', 'trailingPE', 'priceToBook', 'enterpriseToEbitda']]
```

```python
# Calcular multiplos e gerar sinal de valuation
df_val = calcular_multiplos(df_fund)
df_val[['trailingPE', 'priceToBook', 'enterpriseToEbitda',
        'trailingPE_vs_peer', 'priceToBook_vs_peer', 'enterpriseToEbitda_vs_peer',
        'decisao_valuation']]
```

```python
# Heatmap de multiplos vs peer
cols_vs = [c for c in df_val.columns if c.endswith('_vs_peer')]

fig, ax = plt.subplots(figsize=(8, 4))
sns.heatmap(
    df_val[cols_vs].astype(float),
    annot=True, fmt='.2f', cmap='RdYlGn_r',
    center=1.0, linewidths=0.5, ax=ax
)
ax.set_title('Multiplos Relativos vs Mediana do Peer Group\n(>1 = mais caro, <1 = mais barato)')
plt.tight_layout()
plt.show()
```

## Decisao Final

A coluna `decisao_valuation` resume o posicionamento relativo de cada empresa frente ao peer group.
- **barata vs peers**: media dos multiplos relativos < 0.85x
- **neutra vs peers**: entre 0.85x e 1.15x
- **cara vs peers**: > 1.15x
