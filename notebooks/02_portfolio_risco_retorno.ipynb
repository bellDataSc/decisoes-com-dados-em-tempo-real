# 02 - Analise de Risco e Retorno de Portfolio

Baixa precos historicos de uma cesta de ativos, calcula retorno, volatilidade e Sharpe Ratio,
e plota a **fronteira eficiente** via simulacao de Monte Carlo.

**Decisao gerada:** *qual alocacao maximiza retorno ajustado ao risco?*

```python
import sys
sys.path.append('../src')

from data_fetcher import get_historical_prices
from portfolio import otimizar_portfolio, simular_fronteira_eficiente
import matplotlib.pyplot as plt
import pandas as pd
```

```python
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA']
PERIODO  = '3y'
RISK_FREE = 0.05  # taxa Selic aproximada

print('Baixando precos historicos...')
precos = get_historical_prices(TICKERS, period=PERIODO)
precos.tail()
```

```python
# Portfolio de Sharpe Maximo
resultado = otimizar_portfolio(precos, risk_free=RISK_FREE)
print('=== Portfolio Otimo (Max Sharpe) ===')
for k, v in resultado.items():
    print(f'{k}: {v}')
```

```python
# Fronteira Eficiente - Monte Carlo
df_fe = simular_fronteira_eficiente(precos, n_portfolios=4000, risk_free=RISK_FREE)

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(
    df_fe['volatilidade'], df_fe['retorno'],
    c=df_fe['sharpe'], cmap='viridis', alpha=0.5, s=8
)
plt.colorbar(scatter, label='Sharpe Ratio')
# Ponto otimo
ax.scatter(
    resultado['volatilidade_anualizada'],
    resultado['retorno_anualizado'],
    color='red', marker='*', s=200, zorder=5, label='Portfolio Otimo'
)
ax.set_xlabel('Volatilidade Anualizada')
ax.set_ylabel('Retorno Anualizado')
ax.set_title('Fronteira Eficiente - Simulacao de Monte Carlo')
ax.legend()
plt.tight_layout()
plt.show()
```
